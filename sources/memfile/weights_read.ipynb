```python
input_file = "W_PE1.mem"

def hex_to_signed(value: str, bits: int = 8) -> int:
    """
    Convert a hex string to a signed integer using two's complement.
    Example: 'D7' -> -41 for 8-bit values.
    """
    n = int(value, 16)
    if n >= (1 << (bits - 1)):
        n -= (1 << bits)
    return n

```

```python
signed_values = []
with open(input_file, encoding="utf-8") as f:
    for line_number, line in enumerate(f, start=1):
        value = line.strip()
        if not value:
            continue

        try:
            signed_values.append(hex_to_signed(value, bits=8))
        except ValueError:
            raise ValueError(f"Invalid hex value on line {line_number}: {value!r}")

vals = enumerate(signed_values, start=1)  # including line numbers
print(*list(vals)[:784*2], sep='\n')
```
